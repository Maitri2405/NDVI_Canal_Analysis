```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

# Load Data
precip_file = "/content/Narmada_monthly_precipitation.csv"
ndvi_file = "/content/narmada_monthly_ndvi.csv"
lai_file = "/content/Narmada_monthly_LAI_lower.csv"  # Update with actual file name

# Read CSVs
precip_df = pd.read_csv(precip_file)
ndvi_df = pd.read_csv(ndvi_file)
lai_df = pd.read_csv(lai_file)

# Convert 'YearMonth' to datetime format
for df in [precip_df, ndvi_df, lai_df]:
    df["YearMonth"] = pd.to_datetime(df["YearMonth"])
    df["Year"] = df["YearMonth"].dt.year  # Extract year

# Filter data from 2001 onwards
precip_df = precip_df[precip_df["Year"] >= 2001]
ndvi_df = ndvi_df[ndvi_df["Year"] >= 2001]
lai_df = lai_df[lai_df["Year"] >= 2001]

# Scale NDVI from MODIS format (if needed)
ndvi_df["NDVI"] = ndvi_df["NDVI"] / 10000  # Ensure NDVI is in [-1, 1]

# Aggregate Data (Sum for Precipitation, Mean for NDVI & LAI)
yearly_precip = precip_df.groupby("Year")["precip"].sum().reset_index()
yearly_ndvi = ndvi_df.groupby("Year")["NDVI"].mean().reset_index()
yearly_lai = lai_df.groupby("Year")["lai"].mean().reset_index()

# Merge all datasets on Year
df = yearly_precip.merge(yearly_ndvi, on="Year").merge(yearly_lai, on="Year")

# Perform Linear Regression for Trends
def calculate_trend(years, values):
    slope, intercept, r_value, p_value, std_err = linregress(years, values)
    trend_line = slope * years + intercept
    avg_value = np.mean(values)
    trend_percent = (slope / avg_value) * 100 if avg_value != 0 else 0
    return slope, trend_line, trend_percent, p_value

# Compute trends
years = df["Year"]
precip_slope, precip_trend, precip_percent, precip_p = calculate_trend(years, df["precip"])
ndvi_slope, ndvi_trend, ndvi_percent, ndvi_p = calculate_trend(years, df["NDVI"])
lai_slope, lai_trend, lai_percent, lai_p = calculate_trend(years, df["lai"])

# Plot the data
fig, ax1 = plt.subplots(figsize=(10, 5))

# Left Y-axis: Precipitation
ax1.set_xlabel("Year")
ax1.set_ylabel("Total Annual Precipitation (mm)", color="blue")
ax1.plot(years, df["precip"], marker='o', linestyle='-', color="blue", label="Precipitation")
ax1.plot(years, precip_trend, linestyle="dashed", color="darkblue", label=f"Precip Trend ({precip_percent:.2f}%, p={precip_p:.4f})")
ax1.tick_params(axis="y", labelcolor="blue")

# Right Y-axis: NDVI & LAI
ax2 = ax1.twinx()
ax2.set_ylabel("NDVI / LAI", color="green")
ax2.plot(years, df["NDVI"], marker='o', linestyle='-', color="green", label="NDVI")
ax2.plot(years, ndvi_trend, linestyle="dashed", color="darkgreen", label=f"NDVI Trend ({ndvi_percent:.2f}%, p={ndvi_p:.4f})")
ax2.plot(years, df["lai"], marker='s', linestyle='-', color="orange", label="LAI")
ax2.plot(years, lai_trend, linestyle="dashed", color="darkorange", label=f"LAI Trend ({lai_percent:.2f}%, p={lai_p:.4f})")
ax2.tick_params(axis="y", labelcolor="green")

# Legends
fig.suptitle("Yearly Trends of Precipitation, NDVI, and LAI (2001-2023)")
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")

plt.grid()
plt.show()
```
